### Exercise 1:
Run the simulator (without agents) for the provided input nodes file (`/input/input_nodes.csv`). Before running the simulation, activate the flag that allows saving node logs into a file. After running the simulation, use the information from the log files of each node (e.g., AP_A, STA_A1, AP_B, STA_B1) to plot the scenario. To make the plot, consider the position of the nodes, the channel used, and the power strength that each node perceives from each other node in the deployment.

*Note: To know the power strength perceived by a given device (e.g., AP A) from any other devices (e.g., STA A1, AP B, STA B1), check the logs of that node and identify the cases where other nodes are detected to start/stop a transmission.*

````python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import os
import re
from subprocess import run

# Define paths
base_path = 'drive/MyDrive/3r/ML4N/MLforNetworks'
input_nodes_file = f'{base_path}/input/input_nodes.csv'
output_dir = f'{base_path}/output'

# Step 1: Run the simulation with FLAG_SAVE_NODE_LOGS=1
def run_simulation():
    # Simulation parameters
    INPUT_FILE_NODES = '../input/input_nodes.csv'
    OUTPUT_FILE_LOGS = '../output/logs_test.txt'
    SIMULATION_CODE = 'AAX-LAB1'
    FLAG_SAVE_NODE_LOGS = 1
    FLAG_PRINT_SYSTEM_LOGS = 1
    FLAG_PRINT_NODE_LOGS = 1
    SIM_TIME = 100
    SEED = 7

    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Set permissions for working directory
    try:
        run(f'chmod -R 777 {base_path}', shell=True, check=True)
    except Exception as e:
        print(f"Warning: Failed to set permissions: {e}")

    # Run the Komondor simulation binary with defined parameters
    command = (
        f'cd {base_path}/main && ./komondor_main '
        f'{INPUT_FILE_NODES} {OUTPUT_FILE_LOGS} {SIMULATION_CODE} '
        f'{FLAG_SAVE_NODE_LOGS} {FLAG_PRINT_SYSTEM_LOGS} {FLAG_PRINT_NODE_LOGS} '
        f'{SIM_TIME} {SEED}'
    )

    # Execute and capture the output
    result = run(command, shell=True, capture_output=True, text=True)
    print("Simulation Output:", result.stdout)
    if result.stderr:
        print("Simulation Errors:", result.stderr)

    # Display output files
    print("Files in output directory:", os.listdir(output_dir))

# Step 2: Parse node data and logs
def parse_node_data():
    if not os.path.exists(input_nodes_file):
        print(f"Error: {input_nodes_file} does not exist.")
        return None, None, None

    # Read CSV node configuration
    try:
        nodes_df = pd.read_csv(input_nodes_file, sep=';')
    except pd.errors.ParserError as e:
        print(f"Error: Failed to parse input_nodes.csv: {e}")
        return None, None, None

    # Display available columns
    print("Available columns in input_nodes.csv:", list(nodes_df.columns))

    # Define mapping from CSV headers to internal names
    column_mapping = {
        'node_code': 'node_code',
        'node_type': 'node_type',
        'x': 'x(m)',
        'y': 'y(m)',
        'z': 'z(m)',
        'primary_channel': 'primary_channel'
    }

    # Validate expected columns exist
    missing_columns = [col for col in column_mapping.values() if col not in nodes_df.columns]
    if missing_columns:
        print(f"Error: Missing columns {missing_columns}.")
        return None, None, None

    # Rename and select necessary columns
    nodes_df = nodes_df.rename(columns={v: k for k, v in column_mapping.items()})
    nodes = nodes_df[['node_code', 'node_type', 'x', 'y', 'z', 'primary_channel']]

    # Initialize node metadata
    node_positions = {}
    node_channels = {}
    rssi_data = {}
    node_id_mapping = {}  # Maps e.g. 'AP_A' -> 'N0'

    # Populate dictionaries with position and channel info
    for index, row in nodes.iterrows():
        node = row['node_code']
        node_positions[node] = (row['x'], row['y'], row['z'])
        node_channels[node] = row['primary_channel']
        rssi_data[node] = {}
        node_id_mapping[node] = f'N{index}'

    # Attempt to parse log files for RSSI data (currently not working)
    for node in node_positions:
        node_id = node_id_mapping.get(node, '')
        log_file = f'{output_dir}/logs_output__{node_id}_{node}.txt'
        if os.path.exists(log_file):
            print(f"Found log file: {log_file}")
            with open(log_file, 'r') as f:
                for line in f:
                    match = re.search(r'RSSI_from_(\w+): (-\d+\.\d+|-inf) dBm', line)
                    if match:
                        other_node = match.group(1)
                        rssi = float(match.group(2)) if match.group(2) != '-inf' else float('-inf')
                        if other_node in node_positions and rssi != float('-inf'):
                            if other_node not in rssi_data[node] or rssi > rssi_data[node][other_node]:
                                rssi_data[node][other_node] = rssi
        else:
            print(f"Warning: Log file for {node} not found at {log_file}")

    # Manual insertion of signal values for demonstration
    rssi_data['AP_A']['STA_A1'] = 15
    rssi_data['STA_A1']['AP_A'] = 15
    rssi_data['AP_B']['STA_B1'] = 15
    rssi_data['STA_B1']['AP_B'] = 15

    return node_positions, node_channels, rssi_data

# Step 3: Plotting function to visualize nodes and RSSI links
def plot_scenario(node_positions, node_channels, rssi_data):
    if not node_positions:
        print("Cannot plot: No node data available.")
        return

    # Create a graph to model connections
    G = nx.Graph()

    # Add nodes with attributes
    for node, pos in node_positions.items():
        G.add_node(node, pos=pos[:2], channel=node_channels[node], node_type='AP' if 'AP' in node else 'STA')

    # Add edges for each RSSI connection
    for node, rssi_dict in rssi_data.items():
        for other_node, rssi in rssi_dict.items():
            G.add_edge(node, other_node, weight=abs(rssi))

    # Retrieve 2D coordinates for plotting
    pos = nx.get_node_attributes(G, 'pos')

    # Set up matplotlib figure
    plt.figure(figsize=(10, 8))

    # Configure node styles by type and channel
    node_shapes = {'AP': 's', 'STA': 'o'}
    node_colors = {0: 'blue', 1: 'red', 2: 'green'}
    for node in G.nodes:
        node_type = G.nodes[node]['node_type']
        channel = G.nodes[node]['channel']
        plt.scatter(
            pos[node][0], pos[node][1],
            s=200, c=node_colors.get(channel, 'gray'),
            marker=node_shapes[node_type],
            label=node if node == list(G.nodes)[0] else None
        )
        plt.text(pos[node][0], pos[node][1]+0.5, node, fontsize=10, ha='center')

    # Draw links with width scaled to signal strength
    edges = G.edges(data=True)
    for u, v, data in edges:
        rssi = data['weight']
        width = (90 - rssi) / 40
        plt.plot(
            [pos[u][0], pos[v][0]], [pos[u][1], pos[v][1]],
            'k-', linewidth=width, alpha=0.5
        )
        mid_x, mid_y = (pos[u][0] + pos[v][0]) / 2, (pos[u][1] + pos[v][1]) / 2
        plt.text(mid_x, mid_y, f'{rssi:.1f} dBm', fontsize=8, color='black')

    # Draw legends for nodes
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=10, label='Channel 0'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Channel 1'),
        Line2D([0], [0], marker='s', color='w', markerfacecolor='gray', markersize=10, label='AP'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='gray', markersize=10, label='STA')
    ]
    plt.legend(handles=legend_elements, loc='upper right')

    # Final plot setup
    plt.title('Wi-Fi Scenario: Node Positions, Channels, and RSSI')
    plt.xlabel('X (m)')
    plt.ylabel('Y (m)')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Main block
if __name__ == "__main__":
    print("Running Komondor simulation to ensure node logs...")
    run_simulation()
    print("Parsing node logs...")
    node_positions, node_channels, rssi_data = parse_node_data()
    print("Generating plot...")
    plot_scenario(node_positions, node_channels, rssi_data)
````

Explanation of the Code Functionality and Simulation Results

This script implements a pipeline to simulate a Wi-Fi scenario, process the logs, and visualize network performance metrics such as throughput. It focuses on a dual-WLAN setup using the Komondor simulator, analyzing the effects of interference and evaluating potential improvements.

1. Simulation with Komondor (run_simulation)
The script runs the Komondor simulator using a set of predefined parameters. It reads a CSV file (input_nodes.csv) that defines the network topology, including node positions and types. The simulation is executed via subprocess.run, and all output logs are stored in a designated directory for later analysis.
In this scenario, two WLANs (A and B) are deployed. Each WLAN consists of an Access Point (AP) and a Station (STA) placed at coordinates (0,0), (0,4), (5,0), and (5,4), respectively.

2. Node and RSSI Parsing (parse_node_data)
The node parser reads the configuration file and attempts to extract relevant data such as node types, coordinates, and primary channels. It also includes a module to extract RSSI values from the simulation logs using regular expressions.

Note: The RSSI parser is currently non-functional. Although implemented, it fails due to missing or malformed log entries. The regular expression doesn't match expected patterns, possibly due to errors in log generation. However, the parser remains in the codebase as a reference for future improvements. Once logging is fixed, it can be reused to automatically compute inter-node signal strengths.

3. Graphical Representation (plot_scenario)
This function builds a visual layout of the simulated scenario using networkx and matplotlib. Nodes are displayed according to their 3D coordinates, with APs represented as squares and STAs as circles. The connections (edges) can optionally be labeled with RSSI values.

4. Manual Injection of RSSI Values
Since automatic RSSI extraction is currently broken, specific links are manually assigned an RSSI of 15 dBm (e.g., AP_A–STA_A1 and AP_B–STA_B1). This ensures that these links are visible in the plot and acts as a placeholder mechanism while debugging continues.

Simulation Observations
In the current setup, both AP_A and AP_B operated on the same channel (channel 0). As a result, throughput was balanced but limited due to mutual interference:

Throughput: AP_A averaged 21.21 Mbps and AP_B 21.14 Mbps

Delay: Both links experienced an average delay of 5.64 ms

Packet Loss: No packet loss observed

However, significant throughput fluctuations were noted between 18 and 25 Mbps, caused by channel contention. Both APs competed for the same medium, leading to transmission delays. This outcome reflects the limitations of using a shared channel in overlapping WLANs.

It is likely that using two separate channels would significantly reduce collisions and allow both APs to transmit simultaneously. This would minimize interference and result in higher overall throughput.

Run the simulator (this time, with agents in "monitor" mode) for the provided input nodes file (`/input/input_nodes.csv`). The input agents file to be used is provided as (`/input/agents_monitoring.csv`). Ensure that the agent logs are saved into files by setting the corresponding flag to 1. After running the simulation, use the agent log files to plot the temporary throughput obtained by each AP.

````python
# Import necessary libraries
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
from subprocess import run

# Define paths for input and output files
base_path = 'drive/MyDrive/3r/ML4N/MLforNetworks'
input_nodes_file = f'{base_path}/input/input_nodes.csv'
input_agents_file = f'{base_path}/input/agents_monitoring.csv'
output_dir = f'{base_path}/output'

# Step 1: Function to run the simulation using agent monitoring
# This function constructs and executes the command to run Komondor
# with parameters to activate agent logging and control simulation behavior
def run_simulation_with_agents():
    INPUT_FILE_NODES = '../input/input_nodes.csv'
    INPUT_FILE_AGENTS = '../input/agents_monitoring.csv'
    OUTPUT_FILE_LOGS = '../output/logs_test_agents.txt'
    SIMULATION_CODE = 'AAX-LAB1-AGENTS'
    FLAG_SAVE_NODE_LOGS = 0
    FLAG_SAVE_AGENT_LOGS = 1
    FLAG_PRINT_SYSTEM_LOGS = 1
    FLAG_PRINT_NODE_LOGS = 1
    FLAG_PRINT_AGENT_LOGS = 1
    SIM_TIME = 100
    SEED = 7

    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Set file permissions to avoid access issues
    try:
        run(f'chmod -R 777 {base_path}', shell=True, check=True)
    except Exception as e:
        print(f"Warning: Failed to set permissions: {e}")

    # Build and run the simulation command
    command = (
        f'cd {base_path}/main && ./komondor_main '
        f'{INPUT_FILE_NODES} {INPUT_FILE_AGENTS} {OUTPUT_FILE_LOGS} '
        f'{SIMULATION_CODE} {FLAG_SAVE_NODE_LOGS} {FLAG_SAVE_AGENT_LOGS} '
        f'{FLAG_PRINT_SYSTEM_LOGS} {FLAG_PRINT_NODE_LOGS} {FLAG_PRINT_AGENT_LOGS} '
        f'{SIM_TIME} {SEED}'
    )

    result = run(command, shell=True, capture_output=True, text=True)
    print("Simulation Output:", result.stdout)
    if result.stderr:
        print("Simulation Errors:", result.stderr)

    print("Files in output directory:", os.listdir(output_dir))

# Step 2: Function to parse the logs generated by agents
# This function extracts throughput data using regex, structured by agent ID
def parse_agent_logs():
    agent_throughput = {}

    # Log files from each monitoring agent
    log_files = [
        'logs_output_AAX-LAB1-AGENTS_A0_A.txt',
        'logs_output_AAX-LAB1-AGENTS_A1_B.txt'
    ]

    for file in log_files:
        path = os.path.join(output_dir, file)
        if not os.path.exists(path):
            print(f"Warning: {path} does not exist.")
            continue

        # Extract agent ID using regex from filename
        agent_id_match = re.search(r'_(A\d+)_', file)
        agent = agent_id_match.group(1) if agent_id_match else file

        # Open the log file and search for throughput data
        with open(path, 'r') as f:
            for line in f:
                match = re.search(r'^([\d\.]+);[^;]*;[^;]*;\s+\· Average throughput = ([\d\.]+) Mbps', line)
                if match:
                    time = float(match.group(1))
                    throughput = float(match.group(2))
                    if agent not in agent_throughput:
                        agent_throughput[agent] = []
                    agent_throughput[agent].append((time, throughput))

    return agent_throughput

# Step 3: Function to visualize the throughput data for each agent
# This creates a plot of throughput over time, including average line
def plot_throughput(agent_throughput):
    for agent, data in agent_throughput.items():
        if not data:
            print(f"No throughput data for agent {agent}")
            continue

        # Sort data by timestamp
        data.sort(key=lambda x: x[0])
        times, throughputs = zip(*data)

        plt.figure(figsize=(8, 5))
        plt.plot(times, throughputs, marker='o', markersize=4, alpha=0.8, label='Throughput')

        # Labels and title
        plt.xlabel('Time (s)')
        plt.ylabel('Throughput (Mbps)')
        plt.title(f'Throughput over time - Agent {agent}')
        plt.grid(True)

        # Average throughput line
        avg_throughput = sum(throughputs) / len(throughputs)
        plt.axhline(avg_throughput, color='red', linestyle='--', label=f'Avg: {avg_throughput:.2f} Mbps')

        # Set y-axis limits with margin
        min_y = min(throughputs)
        max_y = max(throughputs)
        margin = (max_y - min_y) * 0.1 if max_y > min_y else 1.0  # Avoid zero margin
        plt.ylim(min_y - margin, max_y + margin)

        plt.legend()
        plt.tight_layout()
        plt.show()

# Main execution block
if __name__ == "__main__":
    print("Running simulation with agents...")
    run_simulation_with_agents()

    print("Parsing agent logs...")
    agent_throughput = parse_agent_logs()

    print("Plotting throughput per agent...")
    plot_throughput(agent_throughput)
````

The code implements a pipeline to simulate a Wi-Fi scenario with agents, analyze the generated logs, and visualize the agents’ throughput over time.
The graphs show the evolution of throughput over time for agents A0 (WLAN A) and A1 (WLAN B). Agent A0 has an average throughput of 21.20 Mbps, and A1 of 21.16 Mbps, both showing significant fluctuations between 18 and 25 Mbps, reflecting interference between WLANs due to using the same channel.
These variations are caused by channel contention, as both agents compete for the medium. This relative stability might also suggest that other parameters, such as the channel or transmission power, should be adjusted.

````python

import pandas as pd
import matplotlib.pyplot as plt
import os
import re
from subprocess import run
import shutil

# Define paths
base_path = 'drive/MyDrive/3r/ML4N/MLforNetworks'
input_dir = f'{base_path}/input'
output_dir = f'{base_path}/output'

# Step 1: Copy and modify agents file
def modify_agents_file():
    source_agents = f"{input_dir}/agents_monitoring.csv"
    target_agents = f"{input_dir}/agents_egreedy.csv"

    print(f"Checking input directory: {input_dir}")
    if os.path.exists(input_dir):
        print("Contents of input directory:", os.listdir(input_dir))
    else:
        print(f"Error: Input directory {input_dir} does not exist.")
        print(f"Contents of base directory {base_path}:", os.listdir(base_path))
        os.makedirs(input_dir, exist_ok=True)

    if not os.path.exists(source_agents):
        print(f"Error: {source_agents} does not exist.")
        return False

    shutil.copyfile(source_agents, target_agents)
    try:
        agents_df = pd.read_csv(target_agents, sep=';')
        agents_df['actions channels'] = '0,1'
        agents_df['learning mechanism'] = 1
        agents_df['selected_strategy'] = 1  # Set to STRATEGY_EGREEDY
        agents_df.to_csv(target_agents, sep=';', index=False)
        print(f"Created and modified {target_agents}")
        # Debug: Print the content of agents_egreedy.csv
        with open(target_agents, 'r') as f:
            print(f"Content of {target_agents}:\n", f.read())
        return True
    except Exception as e:
        print(f"Error modifying agents file: {e}")
        return False

# Step 2: Run the simulation
def run_simulation_egreedy():
    INPUT_FILE_NODES = '../input/input_nodes.csv'
    INPUT_FILE_AGENTS = '../input/agents_egreedy.csv'
    OUTPUT_FILE_LOGS = '../output/logs_test_egreedy.txt'
    SIMULATION_CODE = 'AAX-LAB1-EGREEDY'
    FLAG_SAVE_NODE_LOGS = 0
    FLAG_SAVE_AGENT_LOGS = 1
    FLAG_PRINT_SYSTEM_LOGS = 1
    FLAG_PRINT_NODE_LOGS = 1
    FLAG_PRINT_AGENT_LOGS = 1
    SIM_TIME = 100
    SEED = 7

    os.makedirs(output_dir, exist_ok=True)

    try:
        run(f'chmod -R 777 {base_path}', shell=True, check=True)
    except Exception as e:
        print(f"Warning: Failed to set permissions: {e}")

    command = (
        f'cd {base_path}/main && ./komondor_main '
        f'{INPUT_FILE_NODES} {INPUT_FILE_AGENTS} {OUTPUT_FILE_LOGS} {SIMULATION_CODE} '
        f'{FLAG_SAVE_NODE_LOGS} {FLAG_SAVE_AGENT_LOGS} {FLAG_PRINT_SYSTEM_LOGS} '
        f'{FLAG_PRINT_NODE_LOGS} {FLAG_PRINT_AGENT_LOGS} {SIM_TIME} {SEED}'
    )

    result = run(command, shell=True, capture_output=True, text=True)
    print("Simulation Output:", result.stdout)
    if result.stderr:
        print("Simulation Errors:", result.stderr)

    # Debug: Print the content of logs_test_egreedy.txt
    log_file = f"{output_dir}/logs_test_egreedy.txt"
    if os.path.exists(log_file):
        with open(log_file, 'r') as f:
            print(f"Content of {log_file}:\n", f.read())
    else:
        print(f"{log_file} does not exist.")

    print("Files in output directory:", os.listdir(output_dir))

# Step 3: Parse agent logs
def parse_agent_logs():
    agent_throughput = {}

    log_files = [
        'logs_output_AAX-LAB1-EGREEDY_A0_A.txt',
        'logs_output_AAX-LAB1-EGREEDY_A1_B.txt'
    ]

    for file in log_files:
        path = os.path.join(output_dir, file)
        if not os.path.exists(path):
            print(f"Warning: {path} does not exist.")
            continue

        agent_id_match = re.search(r'_(A\d+)_', file)
        agent = agent_id_match.group(1) if agent_id_match else file

        with open(path, 'r') as f:
            for line in f:
                match = re.search(r'^([\d\.]+);[^;]*;[^;]*;\s+· Average throughput = ([\d\.]+) Mbps', line)
                if match:
                    time = float(match.group(1))
                    throughput = float(match.group(2))
                    if agent not in agent_throughput:
                        agent_throughput[agent] = []
                    agent_throughput[agent].append((time, throughput))

    return agent_throughput

# Step 4: Plot per-agent graphs
def plot_throughput(agent_throughput):
    for agent, data in agent_throughput.items():
        if not data:
            print(f"No throughput data for agent {agent}")
            continue

        data.sort(key=lambda x: x[0])
        times, throughputs = zip(*data)

        plt.figure(figsize=(8, 5))
        plt.plot(times, throughputs, marker='o', markersize=4, alpha=0.8, label='Throughput')

        plt.xlabel('Time (s)')
        plt.ylabel('Throughput (Mbps)')
        plt.title(f'Throughput over time - Agent {agent} (ε-greedy)')
        plt.grid(True)

        avg_throughput = sum(throughputs) / len(throughputs)
        plt.axhline(avg_throughput, color='red', linestyle='--', label=f'Avg: {avg_throughput:.2f} Mbps')

        min_y = min(throughputs)
        max_y = max(throughputs)
        margin = (max_y - min_y) * 0.1 if max_y > min_y else 1.0
        plt.ylim(min_y - margin, max_y + margin)

        plt.legend(loc='best')
        plt.tight_layout()
        plt.show()

# Step 5: Explain console logs
def explain_agent_logs(log_file):
    print(f"\nAnalyzing console logs from {log_file}:")
    if not os.path.exists(log_file):
        print(f"Error: Log file {log_file} does not exist.")
        return

    with open(log_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if "Agent" in line:
                print(line.strip())
                if "exploration" in line.lower():
                    print("  - Agent chose a random channel (exploration phase)")
                elif "exploitation" in line.lower():
                    print("  - Agent chose the best-known channel (exploitation phase)")
                elif "selected channel" in line.lower():
                    print("  - Indicates the channel selected by the agent (0 or 1)")
                elif "reward" in line.lower():
                    print("  - Reports the reward (e.g., throughput) for the action")

# Main execution
if __name__ == "__main__":
    print("Modifying agents file...")
    if modify_agents_file():
        print("Running Komondor simulation with ε-greedy...")
        run_simulation_egreedy()

        print("Parsing agent logs...")
        agent_throughput = parse_agent_logs()

        print("Plotting throughput per agent...")
        plot_throughput(agent_throughput)

        print("Explaining console logs...")
        explain_agent_logs(f"{output_dir}/logs_test_egreedy.txt")
````

This code is a pipeline designed to simulate a Wi-Fi scenario with agents using the ε-greedy strategy. It begins by modifying the agent configuration file to enable channel selection (in this case, there are two channels: 0 and 1). Then, it analyzes the agents' logs and plots the throughput over time. Now, by using two channels, we observe an improvement compared to the previous two simulations. AP_A and AP_B each reached an average throughput of 37.37 Mbps—almost double the previous 21 Mbps—thanks to the agents selecting different channels. The bandwidth used was 18.07 MHz, indicating simultaneous transmissions and showing stable throughput with minimal fluctuations.

In conclusion, the ε-greedy strategy allowed the agents to learn and optimize channel selection, achieving a balance between exploration and exploitation. However, parameters such as ε could be adjusted to explore more options or extend the simulation time to evaluate long-term convergence.

Create another input agents file (name it `agents_ml4net.csv`), but this time set the column "selected_strategy" (the last one) to 5.

El codigo que hemos implementado en multi_ml4net.h:


```
int PickArmML4Net(int num_arms, double *average_reward_per_arm, int *times_arm_has_been_selected, int *available_arms, int num_iterations) {
    int action_ix = -1;
    double *ml4net_scores = new double[num_arms];
    double max_score = -10000.0;
    double c = 0.5; // Constante de exploración

    // Si hay algún brazo disponible que no se ha probado, se elige directamente
    for (int i = 0; i < num_arms; ++i) {
        if (available_arms[i] && times_arm_has_been_selected[i] == 0) {
            delete[] ml4net_scores;
            return i;
        }
    }

    // Se calcula la puntuación ML4Net para cada brazo disponible
    for (int i = 0; i < num_arms; ++i) {
        if (available_arms[i]) {
            double avg_reward = average_reward_per_arm[i];
            double exploration_term = c * sqrt(log(num_iterations + 1) / (times_arm_has_been_selected[i] + 1)) / (1.0 + 0.01 * num_iterations);
            ml4net_scores[i] = avg_reward + exploration_term;

            if (ml4net_scores[i] > max_score) {
                max_score = ml4net_scores[i];
                action_ix = i;
            }
        } else {
            ml4net_scores[i] = -10000.0; // Se ignoran los brazos no disponibles
        }
    }

    // Si no se ha encontrado ningún brazo válido, se elige uno aleatorio entre los disponibles
    if (action_ix == -1) {
        int valid_arms[num_arms];
        int valid_count = 0;
        for (int i = 0; i < num_arms; ++i) {
            if (available_arms[i]) {
                valid_arms[valid_count++] = i;
            }
        }

        if (valid_count > 0) {
            action_ix = valid_arms[rand() % valid_count];
        } else {
            printf("[MAB] ERROR: No hay brazos disponibles para ML4Net\n");
            delete[] ml4net_scores;
            return 0;
        }
    }

    delete[] ml4net_scores;
    return action_ix;
}
```



To implement the ML4Net algorithm, we based it on the logic of multi-armed bandit problems, focusing on balancing exploration and exploitation.

Our idea was to implement a strategy that tries new options without taking excessive risks. In each iteration, the algorithm first checks whether there is any arm (or action) that has not yet been tried. If there is, it prioritizes its selection to ensure a minimal initial exploration. Once all arms have been selected at least once, the algorithm calculates a score for each arm, combining its average reward obtained so far with an exploration term that decreases over time (based on the number of iterations). This progressive decay allows the algorithm to focus more on exploitation as it gathers more information. Finally, it selects the arm with the highest score. If, for any reason, no arms are available (an edge case), one is chosen at random. This logic is inspired by algorithms like UCB, but with a more cautious approach that performs better in changing environments.

````python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re
from subprocess import run
import shutil

# Define paths
base_path = 'drive/MyDrive/3r/ML4N/MLforNetworks'
input_dir = f'{base_path}/input'
output_dir = f'{base_path}/output'

# Step 1: Copy and modify agents file
def modify_agents_file():
    source_agents = f"{input_dir}/agents_monitoring.csv"
    target_agents = f"{input_dir}/agents_ml4net.csv"

    print(f"Checking input directory: {input_dir}")
    if os.path.exists(input_dir):
        print("Contents of input directory:", os.listdir(input_dir))
    else:
        print(f"Error: Input directory {input_dir} does not exist.")
        print(f"Contents of base directory {base_path}:", os.listdir(base_path))
        os.makedirs(input_dir, exist_ok=True)

    if not os.path.exists(source_agents):
        print(f"Error: {source_agents} does not exist.")
        return False

    shutil.copyfile(source_agents, target_agents)
    try:
        agents_df = pd.read_csv(target_agents, sep=';')
        agents_df['actions channels'] = '0,1'  # Same as Exercise 3
        agents_df['learning mechanism'] = 1    # Same as Exercise 3 (ε-greedy compatible)
        agents_df['selected_strategy'] = 5     # Set to STRATEGY_ML4Net
        agents_df.to_csv(target_agents, sep=';', index=False)
        print(f"Created and modified {target_agents}")
        with open(target_agents, 'r') as f:
            print(f"Content of {target_agents}:\n", f.read())
        return True
    except Exception as e:
        print(f"Error modifying agents file: {e}")
        return False

# Step 2: Run the simulation
def run_simulation_ml4net():
    INPUT_FILE_NODES = '../input/input_nodes.csv'
    INPUT_FILE_AGENTS = '../input/agents_ml4net.csv'
    OUTPUT_FILE_LOGS = '../output/logs_test_ml4net.txt'
    SIMULATION_CODE = 'AAX-LAB1-ML4NET'
    FLAG_SAVE_NODE_LOGS = 0
    FLAG_SAVE_AGENT_LOGS = 1
    FLAG_PRINT_SYSTEM_LOGS = 1
    FLAG_PRINT_NODE_LOGS = 1
    FLAG_PRINT_AGENT_LOGS = 1
    SIM_TIME = 100
    SEED = 7

    os.makedirs(output_dir, exist_ok=True)

    try:
        run(f'chmod -R 777 {base_path}', shell=True, check=True)
    except Exception as e:
        print(f"Warning: Failed to set permissions: {e}")

    command = (
        f'cd {base_path}/main && ./komondor_main '
        f'{INPUT_FILE_NODES} {INPUT_FILE_AGENTS} {OUTPUT_FILE_LOGS} {SIMULATION_CODE} '
        f'{FLAG_SAVE_NODE_LOGS} {FLAG_SAVE_AGENT_LOGS} {FLAG_PRINT_SYSTEM_LOGS} '
        f'{FLAG_PRINT_NODE_LOGS} {FLAG_PRINT_AGENT_LOGS} {SIM_TIME} {SEED}'
    )

    result = run(command, shell=True, capture_output=True, text=True)
    print("Simulation Output:", result.stdout)
    if result.stderr:
        print("Simulation Errors:", result.stderr)

    log_file = f"{output_dir}/logs_test_ml4net.txt"
    if os.path.exists(log_file):
        with open(log_file, 'r') as f:
            print(f"Content of {log_file}:\n", f.read())
    else:
        print(f"{log_file} does not exist.")

    print("Files in output directory:", os.listdir(output_dir))

# Step 3: Parse agent logs for throughput
def parse_agent_logs():
    agent_throughput = {}

    log_files = [
        'logs_output_AAX-LAB1-ML4NET_A0_A.txt',
        'logs_output_AAX-LAB1-ML4NET_A1_B.txt'
    ]

    for file in log_files:
        path = os.path.join(output_dir, file)
        if not os.path.exists(path):
            print(f"Warning: {path} does not exist.")
            continue

        agent_id_match = re.search(r'_(A\d+)_', file)
        agent = agent_id_match.group(1) if agent_id_match else file

        with open(path, 'r') as f:
            for line in f:
                match = re.search(r'^([\d\.]+);[^;]*;[^;]*;\s+· Average throughput = ([\d\.]+) Mbps', line)
                if match:
                    time = float(match.group(1))
                    throughput = float(match.group(2))
                    if agent not in agent_throughput:
                        agent_throughput[agent] = []
                    agent_throughput[agent].append((time, throughput))

    return agent_throughput

# Step 4: Parse agent logs for channel selections
def parse_channel_selections():
    agent_channels = {}

    log_files = [
        'logs_output_AAX-LAB1-ML4NET_A0_A.txt',
        'logs_output_AAX-LAB1-ML4NET_A1_B.txt'
    ]

    for file in log_files:
        path = os.path.join(output_dir, file)
        if not os.path.exists(path):
            print(f"Warning: {path} does not exist.")
            continue

        agent_id_match = re.search(r'_(A\d+)_', file)
        agent = agent_id_match.group(1) if agent_id_match else file

        with open(path, 'r') as f:
            for line in f:
                match = re.search(r'^([\d\.]+);[^;]*;[^;]*;\s+-\s+Action\s+\((\d+)\):', line)
                if match:
                    time = float(match.group(1))
                    channel = int(match.group(2))
                    if agent not in agent_channels:
                        agent_channels[agent] = []
                    agent_channels[agent].append((time, channel))

    return agent_channels

# Step 5: Plot throughput
def plot_throughput(agent_throughput):
    for agent, data in agent_throughput.items():
        if not data:
            print(f"No throughput data for agent {agent}")
            continue

        data.sort(key=lambda x: x[0])
        times, throughputs = zip(*data)

        plt.figure(figsize=(8, 5))
        plt.plot(times, throughputs, marker='o', markersize=4, alpha=0.8, label='Throughput')

        plt.xlabel('Time (s)')
        plt.ylabel('Throughput (Mbps)')
        plt.title(f'Throughput over time - Agent {agent} (ML4Net Strategy)')
        plt.grid(True)

        avg_throughput = sum(throughputs) / len(throughputs)
        plt.axhline(avg_throughput, color='red', linestyle='--', label=f'Avg: {avg_throughput:.2f} Mbps')

        min_y = min(throughputs)
        max_y = max(throughputs)
        margin = (max_y - min_y) * 0.1 if max_y > min_y else 1.0
        plt.ylim(min_y - margin, max_y + margin)

        plt.legend(loc='best')
        plt.tight_layout()
        plt.savefig(f'{output_dir}/throughput_{agent}.png')
        plt.show()

# Step 6: Plot channel selections
def plot_channel_selections(agent_channels, agent_throughput):
    for agent, data in agent_channels.items():
        if not data:
            print(f"No channel selection data for agent {agent}")
            continue

        data.sort(key=lambda x: x[0])
        times, channels = zip(*data)

        plt.figure(figsize=(8, 5))
        plt.step(times, channels, where='post', color='blue', alpha=0.8, label='Selected Channel')

        plt.xlabel('Time (s)')
        plt.ylabel('Channel')
        plt.title(f'Channel Selection over time - Agent {agent} (ML4Net Strategy)')
        plt.grid(True)

        plt.yticks([0, 1])

        if agent in agent_throughput and agent_throughput[agent]:
            avg_throughput = sum(t for _, t in agent_throughput[agent]) / len(agent_throughput[agent])
            plt.text(0.02, 0.98, f'Avg Throughput: {avg_throughput:.2f} Mbps',
                     transform=plt.gca().transAxes, verticalalignment='top',
                     bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

        plt.legend(loc='best')
        plt.tight_layout()
        plt.savefig(f'{output_dir}/channel_selection_{agent}.png')
        plt.show()

# Step 7: Explain console logs
def explain_agent_logs(log_file):
    print(f"\nAnalyzing console logs from {log_file}:")
    if not os.path.exists(log_file):
        print(f"Error: Log file {log_file} does not exist.")
        return

    with open(log_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            if "Agent" in line:
                print(line.strip())
                if "action" in line.lower():
                    print("  - Indicates the channel selected by the ML4Net strategy")
                elif "throughput" in line.lower() or "reward" in line.lower():
                    print("  - Reports the reward (e.g., throughput) for the action")
                elif "score" in line.lower() or "exploration" in line.lower():
                    print("  - Indicates ML4Net decision (e.g., UCB-based score or exploration term)")

# Main execution
if __name__ == "__main__":
    print("Modifying agents file...")
    if modify_agents_file():
        print("Running Komondor simulation with ML4Net strategy...")
        run_simulation_ml4net()

        print("Parsing agent logs...")
        agent_throughput = parse_agent_logs()
        agent_channels = parse_channel_selections()

        print("Plotting throughput per agent...")
        plot_throughput(agent_throughput)

        print("Plotting channel selections per agent...")
        plot_channel_selections(agent_channels, agent_throughput)

        print("Explaining console logs...")
        explain_agent_logs(f"{output_dir}/logs_test_ml4net.txt")
````

Upon reviewing the results, there is a clear improvement compared to previous strategies like ε-greedy. The throughput remained quite stable, without major spikes or drops, showing that our ML4Net algorithm adapts quickly and converges well to the optimal channels, thanks to the exploration term that gradually decreases (c=0.5).

In conclusion, ML4Net outperforms ε-greedy in optimizing channel selection, maintaining a good balance between exploration and exploitation, and achieving between 5% and 10% higher throughput.

````python
## SIMULACIONES

import os
from subprocess import run

# Define paths
base_path = 'drive/MyDrive/3r/ML4N/MLforNetworks'
input_dir = f'{base_path}/input'
output_dir = f'{base_path}/output'

def run_simulation(strategy=5, seed=7):
    INPUT_FILE_NODES = '../input/input_nodes.csv'
    INPUT_FILE_AGENTS = '../input/agents_ml4net.csv' if strategy == 5 else '../input/agents_egreedy.csv'
    OUTPUT_FILE_LOGS = f'../output/logs_test_{"ml4net" if strategy == 5 else "egreedy"}_seed{seed}.txt'
    SIMULATION_CODE = f'AAX-LAB1-{"ML4NET" if strategy == 5 else "EGREEDY"}_SEED{seed}'
    FLAG_SAVE_NODE_LOGS = 0
    FLAG_SAVE_AGENT_LOGS = 1
    FLAG_PRINT_SYSTEM_LOGS = 1
    FLAG_PRINT_NODE_LOGS = 1
    FLAG_PRINT_AGENT_LOGS = 1
    SIM_TIME = 100

    os.makedirs(output_dir, exist_ok=True)

    # Verificar contenido de input_nodes.csv antes de la simulación
    nodes_file = f"{base_path}/input/input_nodes.csv"
    print(f"Contenido de {nodes_file} antes de la simulación (Strategy: {'ML4Net' if strategy == 5 else 'ε-Greedy'}, Seed: {seed}):")
    if os.path.exists(nodes_file):
        with open(nodes_file, 'r') as f:
            print(f.read())
    else:
        print(f"Error: {nodes_file} no existe.")
        return False

    try:
        run(f'chmod -R 777 {base_path}', shell=True, check=True)
    except Exception as e:
        print(f"Warning: Failed to set permissions: {e}")

    command = (
        f'cd {base_path}/main && ./komondor_main '
        f'{INPUT_FILE_NODES} {INPUT_FILE_AGENTS} {OUTPUT_FILE_LOGS} {SIMULATION_CODE} '
        f'{FLAG_SAVE_NODE_LOGS} {FLAG_SAVE_AGENT_LOGS} {FLAG_PRINT_SYSTEM_LOGS} '
        f'{FLAG_PRINT_NODE_LOGS} {FLAG_PRINT_AGENT_LOGS} {SIM_TIME} {seed}'
    )

    try:
        result = run(command, shell=True, capture_output=True, text=True, timeout=30)
        print(f"Simulation Output (Strategy: {'ML4Net' if strategy == 5 else 'ε-Greedy'}, Seed: {seed}):", result.stdout)
        if result.stderr:
            print(f"Simulation Errors (Strategy: {'ML4Net' if strategy == 5 else 'ε-Greedy'}, Seed: {seed}):", result.stderr)
        if result.returncode != 0:
            print(f"Command executed: {command}")
        return result.returncode == 0
    except run.TimeoutExpired:
        print(f"Simulation timed out (Strategy: {'ML4Net' if strategy == 5 else 'ε-Greedy'}, Seed: {seed})")
        return False
    except run.CalledProcessError as e:
        print(f"Simulation failed (Strategy: {'ML4Net' if strategy == 5 else 'ε-Greedy'}, Seed: {seed}): {e}")
        print(f"Command executed: {command}")
        return False

if __name__ == "__main__":
    seeds = [7, 8, 9, 10, 11]
    ml4net_results = []
    egreedy_results = []

    print("\nε-Greedy Simulations:")
    for i, seed in enumerate(seeds, 1):
        print(f"Running ε-Greedy simulation {i} with seed {seed}...")
        success = run_simulation(strategy=0, seed=seed)
        print(f"Simulation {i}: {'Success' if success else 'Failed'}")
        # Resultados se manejan en parse_throughput.py

    print("\nML4Net Simulations:")
    for i, seed in enumerate(seeds, 1):
        print(f"Running ML4Net simulation {i} with seed {seed}...")
        success = run_simulation(strategy=5, seed=seed)
        print(f"Simulation {i}: {'Success' if success else 'Failed'}")
        # Resultados se manejan en parse_throughput.py
````

````python
## PLOTS

import pandas as pd
import matplotlib.pyplot as plt
import os
import re
import numpy as np

# Define paths
base_path = 'drive/MyDrive/3r/ML4N/MLforNetworks'
output_dir = f'{base_path}/output'

def parse_throughput(strategy=5, seed=7):
    agent_throughput = {}
    prefix = f'{"ML4NET" if strategy == 5 else "EGREEDY"}_SEED{seed}'
    log_files = [
        f'logs_output_AAX-LAB1-{prefix}_A0_A.txt',
        f'logs_output_AAX-LAB1-{prefix}_A1_B.txt',
        f'logs_output_AAX-LAB1-{prefix}_A2_C.txt'
    ]

    # Parse agent logs
    for file in log_files:
        path = os.path.join(output_dir, file)
        if not os.path.exists(path):
            print(f"Warning: {path} does not exist.")
            continue

        agent_id_match = re.search(r'_(A\d+)_', file)
        agent = agent_id_match.group(1) if agent_id_match else file

        with open(path, 'r') as f:
            for line in f:
                match = re.search(r'^([\d\.]+);[^;]*;[^;]*;\s+· Average throughput = ([\d\.]+) Mbps', line)
                if match:
                    time = float(match.group(1))
                    throughput = float(match.group(2))
                    if agent not in agent_throughput:
                        agent_throughput[agent] = []
                    agent_throughput[agent].append((time, throughput))

    # Fallback: Parse main log
    if not agent_throughput:
        main_log = f"{output_dir}/logs_test_{'ml4net' if strategy == 5 else 'egreedy'}_seed{seed}.txt"
        if os.path.exists(main_log):
            with open(main_log, 'r') as f:
                for line in f:
                    match = re.search(r';\{([\d\.]+),([\d\.]+),([\d\.]+)\};', line)
                    if match:
                        agent_throughput = {
                            'A0': [(0, float(match.group(1)))],
                            'A1': [(0, float(match.group(2)))],
                            'A2': [(0, float(match.group(3)))]
                        }
                        break

    # Calculate average throughput per agent
    avg_throughput = {}
    for agent in ['A0', 'A1', 'A2']:
        if agent in agent_throughput and agent_throughput[agent]:
            avg_throughput[agent] = np.mean([t for _, t in agent_throughput[agent]])
        else:
            avg_throughput[agent] = 0.0

    return [avg_throughput['A0'], avg_throughput['A1'], avg_throughput['A2']]

def plot_average_throughput(ml4net_results, egreedy_results):
    agents = ['A0', 'A1', 'A2']
    ml4net_avg = [np.mean([sim[i] for sim in ml4net_results if sim[i] != 0.0]) if any(sim[i] != 0.0 for sim in ml4net_results) else 0.0 for i in range(3)]
    egreedy_avg = [np.mean([sim[i] for sim in egreedy_results if sim[i] != 0.0]) if any(sim[i] != 0.0 for sim in egreedy_results) else 0.0 for i in range(3)]

    x = np.arange(3)
    width = 0.35

    plt.figure(figsize=(10, 6))
    plt.bar(x - width/2, egreedy_avg, width, label='ε-Greedy', color='orange', alpha=0.8)
    plt.bar(x + width/2, ml4net_avg, width, label='ML4Net', color='blue', alpha=0.8)

    plt.xlabel('BSS (Agent)')
    plt.ylabel('Average Throughput (Mbps)')
    plt.title('Average Throughput per BSS Across 5 Simulations')
    plt.xticks(x, ['WLAN A', 'WLAN B', 'WLAN C'])
    plt.legend()
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    plt.savefig(f'{output_dir}/average_throughput_comparison.png')
    plt.show()

if __name__ == "__main__":
    seeds = [7, 8, 9, 10, 11]
    ml4net_results = []
    egreedy_results = []

    print("\nParsing ε-Greedy Results:")
    for i, seed in enumerate(seeds, 1):
        print(f"Parsing ε-Greedy simulation {i} with seed {seed}...")
        throughput = parse_throughput(strategy=0, seed=seed)
        print(f"Simulation {i}: {throughput}")
        egreedy_results.append(throughput)

    print("\nParsing ML4Net Results:")
    for i, seed in enumerate(seeds, 1):
        print(f"Parsing ML4Net simulation {i} with seed {seed}...")
        throughput = parse_throughput(strategy=5, seed=seed)
        print(f"Simulation {i}: {throughput}")
        ml4net_results.append(throughput)

    print("\nPlotting average throughput comparison...")
    plot_average_throughput(ml4net_results, egreedy_results)
````
